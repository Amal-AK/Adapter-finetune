```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import defaultdict

# Suppose df has columns: ['configs','fitness','accuracy','f1','trainable_params_rate']
df = pd.DataFrame(rows)
# 1. Sort the population by fitness
df_sorted = df.sort_values('fitness', ascending=False).reset_index(drop=True)

# 2. Build a usage matrix for the 12 layers
usage_matrix = []
for i in range(len(df_sorted)):
    row_cfg = df_sorted.loc[i, 'configs']  # the 12-element list for that solution
    usage = []
    for layer_idx in range(12):
        usage.append(1 if row_cfg[layer_idx] != 0 else 0)
    usage_matrix.append(usage)

usage_matrix = np.array(usage_matrix)  # shape: (#solutions, 12)

# 3. Plot the heatmap
plt.figure(figsize=(9, 6))
ax = sns.heatmap(
    usage_matrix,
    cmap='Blues',
    cbar=False,
    xticklabels=range(12),
    yticklabels=False
)

plt.xlabel("LLM Layer (0-11)")
plt.ylabel("Solution population(sorted by fitness)")
plt.title("Adapter Usage by Layer (1=Adapter, 0=None)")

# 4. Label rows with the (sorted) row index + fitness
yticks_labels = [
    f"idx={i}"
    for i in range(len(df_sorted))
]

ax.set_yticks(np.arange(len(df_sorted)) + 0.5)  
ax.set_yticklabels(yticks_labels, rotation=0, fontsize=8)

plt.tight_layout()
plt.show()





# Sort and pick top 10 solutions by fitness

df_sorted = df.sort_values('fitness', ascending=False).reset_index(drop=True)
top10 = df_sorted.head(10)

# Define which layers correspond to early, middle, and final
def get_region(layer_idx):
    if layer_idx in [0,1,2,3]:
        return 'early'
    elif layer_idx in [4,5,6,7]:
        return 'middle'
    else:
        return 'final'

# module_usage[ module_name ][ region ] = count
module_usage = defaultdict(lambda: {'early':0, 'middle':0, 'final':0})

for _, row in top10.iterrows():
    configs = row['configs']  # the 12-layer config
    for layer_idx, layer_cfg in enumerate(configs):
        if layer_cfg == 0:
            continue
        region = get_region(layer_idx)
        for mod in layer_cfg['insert_modules']:
            module_usage[mod][region] += 1

# Convert dictionary to a "long" dataframe
rows = []
for mod, counts_dict in module_usage.items():
    rows.append((mod, 'early', counts_dict['early']))
    rows.append((mod, 'middle', counts_dict['middle']))
    rows.append((mod, 'final', counts_dict['final']))

df_plot = pd.DataFrame(rows, columns=['module', 'region', 'count'])
# Pivot so rows = module, columns = region, values = count
df_pivot = df_plot.pivot(index='module', columns='region', values='count').fillna(0)

plt.figure(figsize=(6, 5))

plt.xticks(fontsize=14, rotation=45)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=14, rotation=0) 
sns.set_theme(
    style="whitegrid",        # Background style
    font_scale=1.4        # Scale for all text elements
)

ax = sns.heatmap(
    df_pivot, 
    annot=True,          # show exact counts in each cell
    cmap='Blues',        # or another colormap
    fmt='g',             # 'g' for integer/float output
    cbar_kws={'shrink': 0.7}  # optional styling
)

ax.set_xlabel("Layer Region")
ax.set_ylabel("Insert Module")
ax.set_title("Codebert - clone detection")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt



# 1. Convert Final_population to DataFrame
rows = []
for i, entry in enumerate(Final_population):
    configs = entry[0]  # list of adapter configs (or 0)
    fitness = entry[1]
    acc = entry[2]
    f1 = entry[3]
    params = entry[4]

    rows.append({
        "index": i,
        "configs": configs,
        "fitness": fitness,
        "accuracy": acc,
        "f1": f1,
        "params_rate": params
    })

df = pd.DataFrame(rows)

# 2. Parse adapter configs to extract features
def analyze_config_list(config_list):
    """
    Given a list of adapter configs (or 0), compute:
      - fraction of adapters that have normalization != None
      - mean dropout across all adapters
      - total_count of adapters
    """
    total_adapters = 0
    norm_adapters = 0
    dropout_rates = []

    for adapter_conf in config_list:
        if adapter_conf == 0:
            continue  # means no adapter in that position
        total_adapters += 1

        # Check normalization
        if adapter_conf.get("normalization", None) is not None:
            norm_adapters += 1

        # Collect dropout
        dropout_rates.append(adapter_conf.get("dropout_rate", 0.0))

    fraction_norm = norm_adapters / total_adapters if total_adapters > 0 else 0
    mean_dropout = np.mean(dropout_rates) if total_adapters > 0 else 0.0

    return fraction_norm, mean_dropout, total_adapters

df["frac_norm"] = df["configs"].apply(lambda c: analyze_config_list(c)[0])
df["mean_dropout"] = df["configs"].apply(lambda c: analyze_config_list(c)[1])
df["num_adapters"] = df["configs"].apply(lambda c: analyze_config_list(c)[2])

# Optional: A boolean if there's at least one adapter with normalization
df["has_any_norm"] = df["frac_norm"] > 0



# Example: Sort by 'fitness' descending
df_sorted = df.sort_values(by='fitness', ascending=False).reset_index(drop=True)
print("\nTop 5 Configurations by Fitness:")
print(df_sorted.head(5))

# 4. Correlation Heatmap
plt.figure(figsize=(8, 6))
plt.xticks(fontsize=14, rotation=45)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=14, rotation=0) 
sns.set_theme(
    style="whitegrid",        # Background style
    font_scale=1.8        # Scale for all text elements
)


sns.heatmap(df_sorted[["f1", "params_rate", "frac_norm", "mean_dropout"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Codebert - clone detection")
plt.show()




```
