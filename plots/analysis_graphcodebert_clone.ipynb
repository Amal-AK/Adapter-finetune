```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt



# 1. Convert Final_population to DataFrame
rows = []
for i, entry in enumerate(Final_population):
    configs = entry[0]  # list of adapter configs (or 0)
    fitness = entry[1]
    acc = entry[2]
    f1 = entry[3]
    params = entry[4]

    rows.append({
        "index": i,
        "configs": configs,
        "fitness": fitness,
        "accuracy": acc,
        "f1": f1,
        "params_rate": params
    })

df = pd.DataFrame(rows)

# 2. Parse adapter configs to extract features
def analyze_config_list(config_list):
    """
    Given a list of adapter configs (or 0), compute:
      - fraction of adapters that have normalization != None
      - mean dropout across all adapters
      - total_count of adapters
    """
    total_adapters = 0
    norm_adapters = 0
    dropout_rates = []

    for adapter_conf in config_list:
        if adapter_conf == 0:
            continue  # means no adapter in that position
        total_adapters += 1

        # Check normalization
        if adapter_conf.get("normalization", None) is not None:
            norm_adapters += 1

        # Collect dropout
        dropout_rates.append(adapter_conf.get("dropout_rate", 0.0))

    fraction_norm = norm_adapters / total_adapters if total_adapters > 0 else 0
    mean_dropout = np.mean(dropout_rates) if total_adapters > 0 else 0.0

    return fraction_norm, mean_dropout, total_adapters

df["frac_norm"] = df["configs"].apply(lambda c: analyze_config_list(c)[0])
df["mean_dropout"] = df["configs"].apply(lambda c: analyze_config_list(c)[1])
df["num_adapters"] = df["configs"].apply(lambda c: analyze_config_list(c)[2])

# Optional: A boolean if there's at least one adapter with normalization
df["has_any_norm"] = df["frac_norm"] > 0



# Example: Sort by 'fitness' descending
df_sorted = df.sort_values(by='fitness', ascending=False).reset_index(drop=True)
print("\nTop 5 Configurations by Fitness:")
print(df_sorted.head(5))

# 4. Correlation Heatmap
plt.figure(figsize=(8, 6))
plt.xticks(fontsize=14, rotation=45)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=14, rotation=0) 
sns.set_theme(
    style="whitegrid",        # Background style
    font_scale=1.8        # Scale for all text elements
)
sns.heatmap(df_sorted[["f1", "params_rate", "frac_norm", "mean_dropout"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Graphcodebert - clone detection")
plt.show()






```
